# Data Editing

```python
def extract_year(year):
    year = str(year)
    return year[0:4]
    

df['Start Year']=df['Start Date'].apply(extract_year)
```

# Data Analysis and Visualization

```python
#Male and Female populations.
df['Gender'].value_counts()
```

```python
sns.countplot(x='Gender',data=df,color='#89CFF0')
plt.title('Gender distribution')
```

*The men's population is almost twice the number of female's population.*

```python
#Relation between years and monthly salary.
df.groupby('Years').mean(numeric_only=True)['Monthly Salary']
```

```python
sns.barplot(data=df,x='Years',y='Monthly Salary',estimator=np.average,palette='viridis')
plt.title('Years of experience vs Monthly Salary')
```

*Having more years of experienece doesn't entirely mean having a relatively very high salary compared to other years.*

```python
#Average rate of staying in company.
round(df['Years'].mean(),1)
```

```python
#Most departments in the company.
df['Department'].value_counts()
```

```python
#Most departments in the company categorized by the country.
plt.figure(figsize=(16,9))
sns.countplot(data=df,x='Department',palette='viridis', hue="Country")
plt.xticks(rotation=90)
plt.title('Departments with most employees')
plt.show()
```

```python
#Average and Total monthly salary between female and male employees.
df.groupby('Gender')['Monthly Salary'].agg(['mean', 'sum'])
```

*The average monthly salary between males and females is slightly the same while the total monthly salary is greater for males.*

```python
#Departments with the biggest total monthly salary.
df.groupby('Department').sum(numeric_only=True).sort_values('Monthly Salary',ascending=False)
```

```python
#Box Plot of Annual Salary by Country
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Country', y='Annual Salary')
plt.title("Box Plot of Annual Salary by Country")
```

```python
plt.figure(figsize=(16,9))
plt.xticks(rotation=90)
sns.barplot(x='Department',y='Monthly Salary',data=df,estimator=np.sum,palette='viridis')
plt.xlabel=('Departments')
plt.ylabel('Total Salary')
plt.title("Departments with the most salary")
```

*Obviously the Manufacturing department has the highest total monthly salary and the Research Center has the lowest.*

```python
#Countries with the most salary categorized by gender.
df.groupby(["Country", "Gender"]).sum(numeric_only=True).sort_values('Monthly Salary', ascending=False)['Monthly Salary']
```

```python
plt.figure(figsize=(11,5))
sns.swarmplot(x='Country',y='Monthly Salary',data=df,size=5,hue='Gender',dodge=True)
plt.legend(loc=(1.05,0.5))
plt.title('Swarmplot of Categorized countries vs Monthly salary')
```

*Egypt is the country with the highest total monthly salary.*

```python
#Countries populating the company.
df['Country'].value_counts()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df,x='Country',palette='Set2')
plt.title('Country count')
plt.ylabel('Number of Employees')
```

*The company is mostly populated with Egyptians.*

```python
#Categorized by gender.
sns.catplot(x='Country',data=df,kind='count',col='Gender',palette='Set2')
```

```python
#Centers populating the company.
df['Center'].value_counts()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df,x='Center',palette='Set3')
plt.title('Center count')
plt.ylabel('Number of Employees')
```

*The main center has the highest working employees.*

```python
df['Monthly Salary'].describe().transpose()
```

```python
#Monthly salary analyzation.
sns.set(style='darkgrid')
plt.figure(figsize=(10,6))
sns.displot(x='Monthly Salary',data=df,bins=5,kde=True)
plt.title('Histogram of Monthly Salary')
```

*The most frequent monthly salries among all employees ranges from 2000 to 2250.*

*Aswell, the most common monthly salary ranges from 2000-2500*

```python
#Plotting the KDE of Monthly Salary and comparing it to a normal distribution.
min=df["Monthly Salary"].min()
max=df["Monthly Salary"].max()
sns.kdeplot(x="Monthly Salary",data=df,color='Purple',clip=[min,max],shade=True)
plt.title('KDE of Monthly Salary')
```

```python
#Categorized by gender.
min=df["Monthly Salary"].min()
max=df["Monthly Salary"].max()
sns.kdeplot(x="Monthly Salary",data=df,clip=[min,max],shade=True,hue='Gender')
plt.title('KDE of Monthly Salary based on gender')
```

```python
df['Years'].value_counts()
```

```python
#Years of stay vs Employees count.
sns.countplot(x='Years',data=df,palette="Set2")
plt.title("Years of stay")
plt.ylabel('Number of employees')
```

*As shown here most employees stay for only 1 year, highlighting poor retention and dissatisfaction*

```python
#Job rate across all employees.
df['Job Rate'].value_counts()
```

```python
sns.countplot(x='Job Rate',data=df,palette='viridis')
plt.title('Countplot of Job Rate')
```

*Most job rate given to employees is 5.*

```python
#KDE for Job rate vs Monthly Salary
sns.jointplot(data=df, x='Job Rate', y='Monthly Salary', fill=True, cmap='Greens', alpha=0.3,kind='kde')
```

*Many employees have a job rate of 3 and a salary that's around 2500*


```python
#Overtime hours analyzations.
df['Overtime Hours'].value_counts()
```

```python
min=df["Overtime Hours"].min()
max=df["Overtime Hours"].max()
sns.kdeplot(x='Overtime Hours',data=df,clip=[min,max], shade=True, color='purple')
plt.title('KDE of Overtime hours')
```

*Most overtime hours taken by employees are 10 hours.*

```python
#Highest Average job rates per department.
round(df.groupby('Department').mean(numeric_only=True).sort_values('Job Rate',ascending=False)['Job Rate'],1)
```

```python
plt.figure(figsize=(14,4))
sns.barplot(data=df,x='Department',y='Job Rate',estimator=np.average,palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Departments and Job rates')
plt.show()
```

*As seen here the highest department with average job rates is HR. Which demonstartes how productive it is.*

```python
#Highest average yearly salary for employees per start year.
df.groupby('Start Year').mean(numeric_only=True).sort_values('Annual Salary',ascending=False)['Annual Salary']
```

```python
sns.barplot(data=df,x='Start Year',y='Annual Salary',estimator=np.average,palette='viridis')
plt.title('Barplot of Start Year vs Annual Salary')
```

*Employees that started in 2020 have the highest total average salary out of all employees.*

```python
#Distribution of salaries based on both the employee's gender and center.
sns.catplot(data=df,x='Gender',y='Monthly Salary',kind='violin',col='Center',palette='coolwarm')
```

```python
#Relationship between salaries and job rates.
plt.figure(figsize=(14,5))
sns.swarmplot(data=df, x='Job Rate', y='Monthly Salary')
plt.title('Swarmplot of Job Rate vs Monthly Salary')
```

*It appears that as the job rate increases, the salary does not necessarily rise. As shown in the plot, there are instances where a job rate of 1 has the same salary as a job rate of 5*

```python
#Annual salary distribution.
plt.figure(figsize=(8.5,5))
sns.violinplot(data=df,y='Annual Salary',palette='viridis')
plt.title('Violinplot distribution of Annual salary')
```

*The annual salary median is 2500.*

```python
#Annual salary distribution based on gender.
plt.figure(figsize=(8.5,5))
sns.violinplot(data=df,x='Gender',y='Annual Salary',palette='viridis')
plt.title('Violin plot distribution of annual salary based on gender')
```

*The annual salary median for both male and female employees is slightly the same.*

```python
#Monthly salary distribution over departments using box plot.
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.boxplot(data=df, x="Department", y='Monthly Salary')
plt.title('Box plot of Departments vs Monthly salary')
```

```python
#Monthly salary distribution over departments using violin plot.
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.violinplot(data=df, x="Department", y='Monthly Salary')
plt.title('Violin plot of Departments vs Monthly salary')
```

```python
sns.scatterplot(y='Annual Salary',x='Job Rate',data=df)
plt.title('Scatterplot of Job rate vs Monthly salary')
```

*Some employees with the same job rate receive the same monthly salary(1 and 5). This means that the salary doesn't depend on job performance*

```python
#Job rate distribuition by gender.
sns.boxplot(data=df, x='Gender', y='Job Rate',palette='Set2')
plt.title("Violin plot distribution of Job Rate based on gender")
```
